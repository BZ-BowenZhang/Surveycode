```python
location_point = (51.523884, -0.1349691)
```

```python
G = ox.graph_from_point(location_point, distance=1000, distance_type='bbox', network_type='walk')
```

```python
G2 = ox.project_graph(G)
```

```python
nodes=ox.graph_to_gdfs(G2, nodes=True, edges=False)
```

```python
edges = ox.graph_to_gdfs(G2, nodes=False, edges=True)
```

```python
survey_data=gpd.read_file("Saved_data.geojson")
```

```python
df=pd.read_csv("std.csv")
```

```python
edges["poi"]=df["poi"]
edges["crime"]=df["crime"]
edges["poi_std"]=df["poi_std"]
edges["crime_std"]=df["crime_std"]
edges["trees"]=df['trees']
edges["greenness_std"]=df["greenness_std"]
```

```python
edges.to_csv('output.csv')
```

```python
a=float(input("percentage of Safty\n"))/100
b=float(input("percentage of Green\n"))/100
c=float(input("percentage of Social\n"))/100
if((math.fabs(a+b+c-1))>0.00001):
    print("Warning (TOTAL WEIGHT not equal to 1): Please check the weight, it may cause the algorithm unvalied")
else:
    print("Weight Check Passed")

edges["W"]=edges["length"]*(1+a*edges["crime_std"]-b*edges["poi_std"]-c*edges["greenness_std"])
G=ox.save_load.gdfs_to_graph(nodes,edges)
```

```python
o = utm.from_latlon(51.5246171,-0.1425432)
d = utm.from_latlon(51.5230848,-0.1265042)
orig_xy = (o[1], o[0])
target_xy = (d[1], d[0])
orig_node = ox.get_nearest_node(G, orig_xy, method='euclidean')
target_node = ox.get_nearest_node(G, target_xy, method='euclidean')
o_closest = nodes.loc[orig_node]
t_closest = nodes.loc[target_node]
od_nodes = gpd.GeoDataFrame([o_closest, t_closest], geometry='geometry', crs=nodes.crs)
route_L = nx.shortest_path(G=G, source=orig_node, target=target_node, weight='length')
route_W = nx.shortest_path(G=G, source=orig_node, target=target_node, weight='W')
route_L_nodes = nodes.loc[route_L]
route_L_nodes['Group_Number']="L"
route_W_nodes = nodes.loc[route_W]
route_W_nodes['Group_Number']="W"
route_line_L=route_L_nodes.groupby(['Group_Number'])['geometry'].apply(lambda x: LineString(x.tolist()))
route_line_W=route_W_nodes.groupby(['Group_Number'])['geometry'].apply(lambda x: LineString(x.tolist()))
gdf_W_for_show=gpd.GeoDataFrame(geometry=route_line_W,crs=edges.crs)
gdf_L_for_show=gpd.GeoDataFrame(geometry=route_line_L,crs=edges.crs)
gdf_W_for_show_wgs_84=gdf_W_for_show.to_crs("+init=epsg:4326")
gdf_L_for_show_wgs_84=gdf_L_for_show.to_crs("+init=epsg:4326")
gdf_W_for_show_wgs_84.to_file("route_w.geojson",driver="GeoJSON")
gdf_L_for_show_wgs_84.to_file("route_L.geojson",driver="GeoJSON")
```

```python
gdf_L_for_show['survey time']=time.now()
gdf_L_for_show['origin']=Point(o[0],o[1])
gdf_L_for_show["destination"]=Point(d[0],d[1])
gdf_L_for_show['route type']="L"
gdf_W_for_show['survey time']=time.now()
gdf_W_for_show['origin']=Point(o[0],o[1])
gdf_W_for_show["destination"]=Point(d[0],d[1])
gdf_W_for_show['route type']="W"
satisfied=int(input("please input the satisfication of this route in range(1-5)"))
gdf_W_for_show['satisfication']=satisfied
gdf_L_for_show['satisfication']=-1
survey_data=survey_data.append(gdf_W_for_show)
survey_data=survey_data.append(gdf_L_for_show)
```
