```python
poi=pd.read_csv('cleaned_poi-2.csv')
```

```python
geometry=[Point(xy) for xy in zip(poi.x, poi.y)]
```

```python
crs_bng={'init':"epsg:27700"}
```

```python
geo_poi=gpd.GeoDataFrame(poi,crs=crs_bng,geometry=geometry)
```

```python
geo_poi.to_crs(buffer_edges.crs)
```

```python
pts = geo_poi.copy() 
```

```python
pts=pts.to_crs("+proj=utm +zone=30 +ellps=WGS84 +datum=WGS84 +units=m +no_defs")
```

```python
pts
```

```python
pts_in_polys = []

# Loop over polygons with index i.
for i, poly in buffer_edges.iterrows():

    # Keep a list of points in this poly
    pts_in_this_poly = []

    # Now loop over all points with index j.
    for j, pt in pts.iterrows():
        if poly.geometry.contains(pt.geometry):
            # Then it's a hit! Add it to the list,
            # and drop it so we have less hunting.
            pts_in_this_poly.append(pt.geometry)

    # We could do all sorts, like grab a property of the
    # points, but let's just append the number of them.
    pts_in_polys.append(len(pts_in_this_poly))

# Add the number of points for each poly to the dataframe.
buffer_edges['number of points'] = gpd.GeoSeries(pts_in_polys)
```

```python
x1,y1,z1,zl1=utm.from_latlon(51.5238865,-0.136504)
```

```python
casa = Point (x1, y1)
```

```python
rs_station= Point (699370.13, 5711840.86)
```

```python
print(casa)
```

```python
print(rs_station)
```

```python
orig_xy = (casa.y, casa.x)
```

```python
target_xy = (rs_station.y, rs_station.x)
```

```python
orig_node = ox.get_nearest_node(graph_proj, orig_xy, method='euclidean')
```

```python
target_node = ox.get_nearest_node(graph_proj, target_xy, method='euclidean')
```

```python
o_closest = nodes_proj.loc[orig_node]
```

```python
t_closest = nodes_proj.loc[target_node]
```

```python
print(orig_node)
```

```python
print(target_node)
```

```python
od_nodes = gpd.GeoDataFrame([o_closest, t_closest], geometry='geometry', crs=nodes_proj.crs)
```

```python
od_nodes
```

```python
graph_proj
```

```python
route = nx.shortest_path(G=graph_proj, source=orig_node, target=target_node, weight='length')
```

```python
print(graph_proj)
```

```python
fig, ax = ox.plot_graph_route(graph_proj, route, origin_point=orig_xy, destination_point=target_xy)
```

```python
plt.tight_layout()
```

```python
route_nodes = nodes_proj.loc[route]
```

```python
route_nodes['Number']=1
```

```python
route_nodes
```

```python
route_line=route_nodes.groupby(['Number'])['geometry'].apply(lambda x: LineString(x.tolist()))
```

```python
route_line_df = gpd.GeoDataFrame(geometry=route_line)
```
