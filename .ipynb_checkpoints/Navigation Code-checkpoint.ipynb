```python
casa= Point (698640.5643363819, 5711973.088101764)
```

```python
rs_station= Point (699370.13, 5711840.86)
```

```python
orig_xy = (casa.y, casa.x)
```

```python
target_xy = (rs_station.y, rs_station.x)
```

```python
orig_node = ox.get_nearest_node(G, orig_xy, method='euclidean')
```

```python
target_node = ox.get_nearest_node(G, target_xy, method='euclidean')
```

```python
nodes_proj, edges_proj = ox.graph_to_gdfs(G, nodes=True, edges=True)
```

```python
o_closest = nodes_proj.loc[orig_node]
```

```python
t_closest = nodes_proj.loc[target_node]
```

```python
od_nodes = gpd.GeoDataFrame([o_closest, t_closest], geometry='geometry', crs=nodes_proj.crs)
```

```python
route = nx.shortest_path(G=G, source=orig_node, target=target_node, weight='length')
```

```python
fig, ax = ox.plot_graph_route(G, route, origin_point=orig_xy, destination_point=target_xy)
```
